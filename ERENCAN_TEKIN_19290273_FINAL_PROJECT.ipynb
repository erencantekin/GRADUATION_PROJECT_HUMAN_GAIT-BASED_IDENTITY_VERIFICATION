```python
import mediapipe as mp
import cv2

mDrawing = mp.solutions.drawing_utils
mHolistic = mp.solutions.holistic
```

```python
import os
import numpy as np
import csv
```

```python
# find the max number of pose coordinats of the image to create model

cap = cv2.VideoCapture("class1.mp4")
with mHolistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:

    while cap.isOpened():
        isRead, frames = cap.read() # read frames, if it is successfull isRead is true

        # openCV iamges are BGR format by default, convert them to RGB so holisticm odel can use
        image = cv2.cvtColor(frames, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False

        # start detecting
        detectedCoordinats = holistic.process(image)

        # convert back from RGB to BGR 
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

        # Left Hand Coordinats
        mDrawing.draw_landmarks(image, detectedCoordinats.left_hand_landmarks, mHolistic.HAND_CONNECTIONS,
                                mDrawing.DrawingSpec( color = (121, 22, 76), thickness = 2, circle_radius= 4),
                                mDrawing.DrawingSpec( color = (121, 44, 250), thickness = 2, circle_radius = 2)
                               )

        # Right Hand Coordinats
        mDrawing.draw_landmarks(image, detectedCoordinats.right_hand_landmarks, mHolistic.HAND_CONNECTIONS,
                                mDrawing.DrawingSpec( color = (121, 22, 76), thickness = 2, circle_radius= 4),
                                mDrawing.DrawingSpec( color = (121, 44, 250), thickness = 2, circle_radius = 2)
                               )

        # Pose Detections
        mDrawing.draw_landmarks(image, detectedCoordinats.pose_landmarks, mHolistic.POSE_CONNECTIONS,
                                mDrawing.DrawingSpec( color = (245, 117, 66), thickness = 2, circle_radius=4),
                                mDrawing.DrawingSpec( color = (245, 66, 230), thickness = 2, circle_radius = 2)
                               )

        

        cv2.imshow("Webcam",image)
        if cv2.waitKey(1):  # & 0xFF == ord('e'):
            cap.release()
            cv2.destroyAllWindows()
            break
        

maxNumberOfPoseCoordinats = len(detectedCoordinats.pose_landmarks.landmark)
```

```python
# create CSV file
with open( "ClassCoordValues.csv", mode = 'w', newline = "") as csvFile:
    writer = csv.writer(csvFile, delimiter = ",", quotechar = '"', quoting= csv.QUOTE_MINIMAL)
    writer.writerow(modelLandmarks)
    
```

```python
cap = cv2.VideoCapture(imagePath)
with mHolistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:

    while cap.isOpened():
        isRead, frames = cap.read() # read frames, if it is successfull isRead is true

        if(isRead ):
            # openCV iamges are BGR format by default, convert them to RGB so holisticm odel can use
            image = cv2.cvtColor(frames, cv2.COLOR_BGR2RGB)
            image.flags.writeable = False
    
            # start detecting
            detectedCoordinats = holistic.process(image)
    
            # convert back from RGB to BGR 
            image.flags.writeable = True
            image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
    
            # Left Hand Coordinats
            mDrawing.draw_landmarks(image, detectedCoordinats.left_hand_landmarks, mHolistic.HAND_CONNECTIONS,
                                    mDrawing.DrawingSpec( color = (121, 22, 76), thickness = 2, circle_radius= 4),
                                    mDrawing.DrawingSpec( color = (121, 44, 250), thickness = 2, circle_radius = 2)
                                   )
    
            # Right Hand Coordinats
            mDrawing.draw_landmarks(image, detectedCoordinats.right_hand_landmarks, mHolistic.HAND_CONNECTIONS,
                                    mDrawing.DrawingSpec( color = (121, 22, 76), thickness = 2, circle_radius= 4),
                                    mDrawing.DrawingSpec( color = (121, 44, 250), thickness = 2, circle_radius = 2)
                                   )
    
            # Pose Detections
            mDrawing.draw_landmarks(image, detectedCoordinats.pose_landmarks, mHolistic.POSE_CONNECTIONS,
                                    mDrawing.DrawingSpec( color = (245, 117, 66), thickness = 2, circle_radius=4),
                                    mDrawing.DrawingSpec( color = (245, 66, 230), thickness = 2, circle_radius = 2)
                                   )
            
            # Extract Pose Coordinats Landmarks
            try:
                
                extractedLandmarks = detectedCoordinats.pose_landmarks.landmark
                extractedLandmarksRow = list(np.array( [ [ poseL.x, poseL.y, poseL.z, poseL.visibility ] for poseL in extractedLandmarks ] ).flatten() )

                extractedLandmarksRow.insert(0, className)

                with open("ClassCoordValues.csv", mode = "a", newline = "") as csvFile:
                    writer = csv.writer(csvFile, delimiter = ",", quotechar='"', quoting = csv.QUOTE_MINIMAL)
                    writer.writerow(extractedLandmarksRow)

                    #print values
                    print(extractedLandmarksRow)
            except:
                # detection interrupt
                pass
        

            cv2.imshow("Webcam",image)
            if cv2.waitKey(1) & 0xFF == ord('e'):
                cap.release()
                cv2.destroyAllWindows()
                break
                
        else: # image/video is end
            cap.release()
            cv2.destroyAllWindows()
            break
```

```python
# Train Model with python module named scikit-learn
from sklearn.model_selection import train_test_split
import pandas as pd

coordFile = pd.read_csv('ClassCoordValues.csv')
```

```python
# head
#coordFile.head()

#specific class
coordFile[ coordFile['class'] == 'class1' ]

# tail
#coordFile.tail()
```

```python

# features
feature_dataset = coordFile.drop('class', axis = 1 )

# target values
target_dataset = coordFile['class']

# seperate datas into train and test

trainX, testX, trainY, testY = train_test_split( feature_dataset, target_dataset, test_size = 0.3, random_state = 1234 ) # 30% testing,  70% training
```

```python
# show the results
trainX
#testY
```

```python
# train model using pipelines ( classification )

from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression, RidgeClassifier


pipelines = {
    "rf": make_pipeline(StandardScaler(), RandomForestClassifier()),
    "gb": make_pipeline(StandardScaler(), GradientBoostingClassifier()),
    "rc": make_pipeline(StandardScaler(), RidgeClassifier()),
    "lr": make_pipeline(StandardScaler(), LogisticRegression(solver="saga", max_iter=5000)),
}
```

```python
# drop rows with missing values

trainX = trainX.dropna()
trainY = trainY.loc[trainX.index]


fitModel = {}
for algo, pipeline in pipelines.items():
    model = pipeline.fit(trainX.values, trainY.values)
    fitModel[algo] = model
```

```python
# drop rows with missing values

testX = testX.dropna()
testY = testY.loc[testX.index]

# ridge classifier
fitModel["rc"].predict(testX.values)
```

```python
# accuracy & Serialization
from sklearn.metrics import accuracy_score
import pickle


for algo, model in fitModel.items():
    predictions = model.predict(testX.values) # hold predictions
    # print accuracy
    print(algo, accuracy_score(testY, predictions))
```

```python
fitModel["rf"].predict(testX.values)
```

```python
# serialization

with open("ClassCoordValues.pkl", "wb") as pickleFile:
    pickle.dump(fitModel["rf"], pickleFile)
```

```python
# test model

# deserialization
with open("ClassCoordValues.pkl", "rb") as pickleFile:
    binaryModel = pickle.load(pickleFile)
```
